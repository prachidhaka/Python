```python
# 4. Calculate the following from merged dataset:
# a. Time Period of the available transaction data.
```

```python

# Printing a sample of the 'tran_date' column
print(csf['tran_date'].head())
```

```python
# Convert the 'tran_date' column to datetime
csf['tran_date'] = pd.to_datetime(csf['tran_date'], format='%d-%m-%Y', dayfirst=True, errors='coerce')
#'errors='coerce'' replaces invalid dates with NaT.

# Display any rows with 'NaT' values

display(csf[csf['tran_date'].isna()])

# Calculate the time period of the valid transaction data
valid_dates = csf[csf['tran_date'].notna()]['tran_date']
display(valid_dates.max() - valid_dates.min())

#Helps identify seasonal trends and forecast demand.
```

```python
# b. Count of transactions where the total amount of transaction was negative
csf.total_amt.iloc[(csf.total_amt > 0).values].count()
#(csf.total_amt > 0).values generates a boolean array
#and .iloc selects rows based on the boolean array.
```

```python

# Grouping data by 'Gender' and 'prod_cat', and counting the number of transactions
popularity = csf.groupby(['Gender', 'prod_cat']).size().unstack(fill_value=0)

# Displaying the resulting DataFrame
popularity
```

```python
# 6. Which City code has the maximum customers and what was the percentage of customers from that city?


csf.groupby('city_code')['transaction_id'].count().reset_index().sort_values(by='transaction_id',ascending=False)
```

```python
# Percentage of customers
(csf.groupby('city_code')['transaction_id'].count().reset_index().sort_values(by='transaction_id',ascending=False).iloc[0,1]/csf['transaction_id'].count())*100
```

```python
# 7. Which store type sells the maximum products by value and by quantity?

csf.groupby('Store_type')[['total_amt','Qty']].sum().reset_index().sort_values(by='total_amt',ascending=False).head(1)
```

```python
# 8. What was the total amount earned from the "Electronics" and "Clothing" categories from Flagship Stores?

csf.total_amt.loc[(csf['prod_cat']=='Electronics')|(csf['prod_cat']=='Clothing')].sum()
```

```python
# 9. What was the total amount earned from "Male" customers under the "Electronics" category?

csf.loc[(csf['prod_cat']=='Electronics')|(csf['Gender']=='Male')].total_amt.sum()
```

```python
# 10. How many customers have more than 10 unique transactions, after removing all transactions which have any negative amounts?
# Filtering customers that have only positive total_amt i.e. Sales completed without returns.
csf_sales=csf.loc[csf.total_amt>0]


top_tran=csf_sales.groupby('cust_id')['transaction_id'].count().reset_index().sort_values(by='transaction_id',ascending=False)
(top_tran['transaction_id']>10).sum()
```

```python
filtered_data = customer_final[customer_final['total_amt'] >= 0]
customers_with_more_than_10_transactions = filtered_data['cust_id'].value_counts()[filtered_data['cust_id'].value_counts() > 10].count()
customers_with_more_than_10_transactions
```

```python
# 11. For all customers aged between 25 - 35, find out:
# Getting Ages for customers from DOB to 1st Mar, 2014.
csf['DOB']=pd.to_datetime(csf['DOB'])
dateforage=pd.to_datetime('2014-3-1')
csf['Age']=((dateforage - csf['DOB']).dt.total_seconds()/(365.25*24*3600)).astype('int')

# Filtering by ages 25 to 35.
csf_25_35=csf.loc[(csf.Age > 25)&(csf.Age <35)]
# a. What was the total amount spent for “Electronics” and “Books” product categories?
csf_25_35_EB=csf_25_35.loc[(csf_25_35.prod_cat =='Electronics')|(csf_25_35.prod_cat =='Books')]
display(csf_25_35_EB.agg({'total_amt':'sum'}).reset_index())

# b. What was the total amount spent by these customers between 1st Jan, 2014 to 1st Mar, 2014?
filter = ((csf['tran_date'] >= '1st Jan, 2014')&(csf['tran_date'] <= '1st Mar, 2014'))
display(csf['total_amt'].loc[filter].sum())
```
